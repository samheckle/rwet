# trying to do some blackout poetry

inspired by blackout poetry, i tried a quick experiement with removing words but keeping the spacing. this doesn't so much fiddle with the structure of the text but rather which text is perceived. 

```python
import random
import textwrap

for line in open("tale_of_benjamin_bunny.txt"):
    words = line.split() 
    output = [] 
    sel = random.sample(words, random.randint(0,len(words)))
    
    for item in words:
        if item in sel:
            output.append(item)
        for char in item:
            output.append(' ')
    print(textwrap.fill(''.join(output), 60))
```
