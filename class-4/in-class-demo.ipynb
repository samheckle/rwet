```python
# [predicate-expression for temp-var-name in source-list if membership-expression]
# predicate-expression = for every item in the list
fives = [item for item in words if len(item) == 5]
```

```python
[item for item in fives if item[0]== 'c' and 'n' in item]
```

```python
words = [item for item in words if len(item) > 0]
```

```python
a_list = [item for item in words if item[0] == 'a']
l_list = [item for item in words if item[0] == 'l']
i_list = [item for item in words if item[0] == 'i']
s_list = [item for item in words if item[0] == 's']
o_list = [item for item in words if item[0] == 'o']
n_list = [item for item in words if item[0] == 'n']
```

```python
print(random.choice(a_list).title())
print(random.choice(l_list).title())
print(random.choice(l_list).title())
print(random.choice(i_list).title())
print(random.choice(s_list).title())
print(random.choice(o_list).title())
print(random.choice(n_list).title())
```
